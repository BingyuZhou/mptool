```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def road_represent(clothoid_path, represent_path):
    plt.figure(figsize=(10,8))
    clothoid = np.loadtxt(clothoid_path)
    plt.plot(clothoid[:,0], clothoid[:,1], '.-', color='g', label='true')

    bary = np.loadtxt(represent_path)

    plt.plot(bary[:,0], bary[:,1], '*-', color='r', label='bary-repre')
    plt.grid()
    plt.legend()
    plt.axis('equal')

clothoid_path = 'build/unit_tests/clothoid.csv'
represent_path = 'build/unit_tests/ref_path.csv'

road_represent(clothoid_path, represent_path)
```
